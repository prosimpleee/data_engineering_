```python
from pyspark.sql import SparkSession
import pyspark.sql.functions as F
from pyspark.sql.functions import *
import datetime
from pyspark.sql.types import IntegerType
spark = SparkSession.builder.getOrCreate()
```

```python
spark\
.read\
.option('header',True)\
.option('inferSchema', True)\
.csv('/home/jovyan/companies_sorted.csv')\
.limit(300)\
.createOrReplaceTempView('company_dataset')
```

```python
#Create df
df = spark.table('company_dataset')
df.toPandas()
```

```python
df = df\
.withColumn('year_founded', (F.col('year founded')).cast(IntegerType()))\
.withColumn('company_name', F.initcap(F.col('name')))\
.withColumn('industry', F.initcap(F.col('industry')))\
.withColumn('country', F.initcap(F.col('country')))\
.withColumn('linkedin_url', F.col('linkedin url'))\
.withColumn('total_employees', F.col('total employee estimate').cast(IntegerType()))
```

```python
df = df\
.orderBy('total_employees', ascending = False)
```

```python
df.dtypes
```

```python
df.limit(10).toPandas()
```

```python
df = df.dropna()
```

```python
df = df\
.select(F.col('company_name'),
        F.col('country'),
        F.col('industry'),
        F.col('year_founded'),
        F.col('domain'),
        F.col('total_employees'),
        F.col('linkedin_url'))
```

```python
df.limit(10).toPandas()
```

```python
df_avg_total_employees_count = df\
.groupBy('country')\
.agg(
    F.avg(F.col('total_employees').alias('avg_tot_employees')).alias('avg_count')
        )
```

```python
df_avg_total_employees_count = df_avg_total_employees_count\
.withColumn('avg_tot_employees', F.col('avg_count').cast(IntegerType()))
```

```python
df_avg_total_employees_count.select(F.col('country'), F.col('avg_tot_employees'))\
.orderBy('avg_tot_employees', ascending = False)\
.toPandas()
```
